```python
import pandas as pd
import numpy as np;
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../Data collecting/phone_trendyol_data.csv")
```

```python
data.info()
```

```python
data = data.drop(["Ekran Boyutu Aralığı","Ekran Boyutu",'Suya/Toza Dayanıklılık','Çift Hat','Bağlantılar','Radio','NFC',"Ön Kamera Çözünürlüğü","Ana Kamera Çözünürlül Aralığı","Batarya Kapasitesi Aralığı","CPU Aralık","Ön Kamera Çözünürlük Aralığı","Şarj Hızı","Ekran Boyut Aralığı"], axis=1)
```

```python
data.columns
```

```python
data
```

```python
#Hangi modelden kaç adet telefon var onu gösteriyor
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x='Cep Telefonu Modeli',data = data,order = data['Cep Telefonu Modeli'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

```python
#TL Yazısını kaldırıyoruz
for i in range(0,24):
    data["prices"] = data.prices.str.replace("TL","")
```

```python
data
```

```python
#virgülleri kaldırıyoruz
for i in range(0,100):
    data["prices"] = data.prices.str.replace(',{}'.format(i), '')

data["prices"] = data.prices.str.replace(',00'.format(i), '')
data["prices"] = data.prices.str.replace(',01'.format(i), '')
data["prices"] = data.prices.str.replace(',02'.format(i), '')
data["prices"] = data.prices.str.replace(',03'.format(i), '')
data["prices"] = data.prices.str.replace(',04'.format(i), '')
data["prices"] = data.prices.str.replace(',05'.format(i), '')
data["prices"] = data.prices.str.replace(',06'.format(i), '')
data["prices"] = data.prices.str.replace(',07'.format(i), '')
data["prices"] = data.prices.str.replace(',08'.format(i), '')
data["prices"] = data.prices.str.replace(',09'.format(i), '')
data["prices"] = data.prices.str.replace('.', '')
```

```python
data['prices'] = data.prices.astype(float)
```

```python
data["prices"]
```

```python
data.head()
```

```python
#Türkçe karakterleri ingilizce karakterine dönüştürüyoruz
data=data.rename(columns = {'Arka Kamera Çözünürlüğü':'Arka_Kamera_Cozunurlugu','Bellek Kapasitesi':'Bellek_Kapasitesi','Batarya Kapasitesi':'Batarya_Kapasitesi','Ekran Boyutu':'Ekran_Boyutu','Ekran Çözünürlüğü':'Ekran_Cozunurlugu',"Dahili Hafıza":"Dahili_Hafiza","Video Kayıt Çözünürlüğü":"Video_Kayit_Cozunurlugu","Arka Kamera Sayısı":"Arka_Kamera_Sayisi","Ekran Yenileme Hızı":"Ekran_Yenileme_Hizi","Ön Kamera Sayısı":"On_Kamera_Sayisi"})
```

```python
data.head()
```

```python
#GB yazısını kaldırıyoruz
data["Dahili_Hafiza"] = data.Dahili_Hafiza.str.replace('GB', '')
```

```python
#Hz yazısını kaldırıyoruz
data["Ekran_Yenileme_Hizi"] = data.Ekran_Yenileme_Hizi.str.replace('Hz', '')
```

```python
data
```

```python
#Sadece sayı olan değerleri object türünden float türüne dönüştürüyoruz
data["Dahili_Hafiza"] = data.Dahili_Hafiza.astype(float)
data["Ekran_Yenileme_Hizi"] = data.Ekran_Yenileme_Hizi.astype(float)
```

```python
#İşletim sistemine göre ürün sayısı 
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x='İşletim Sistemi',data = data,order = data['İşletim Sistemi'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```
